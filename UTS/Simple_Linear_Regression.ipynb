```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
plt.scatter(dataset.iloc[:, 3], dataset.iloc[:, 11])
plt.title('Scatter Plot of Data')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.show()
```

```python
# Memeriksa korelasi antar fitur dengan heatmap
import seaborn as sns
correlation_matrix = dataset.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Splitting the dataset into independent (X) and dependent (y) variables
X = dataset.iloc[:, 3:4].values
y = dataset.iloc[:, 11:12].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Printing the shapes of the training and test sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Train the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Test the model
y_pred = regressor.predict(X_test)
```

```python
# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("R-squared (R2) Score:", r2)
```

```python
import joblib
joblib.dump(regressor, 'linear_regression_model.pkl')
```

```python
# Optional: Visualize the training set and the regression line
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_train, regressor.predict(X_train), color='red', linewidth=2, label='Regression Line')
plt.title('Simple Linear Regression - Training Set')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.show()
```
